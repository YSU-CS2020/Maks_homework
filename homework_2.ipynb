Question 24. Rewrite match/case into if/elif/else

```python
"""
match color:
    case "red":
        meaning = "Stop"
    case "yellow":
        meaning = "Wait"
    case "green":
        meaning = "Go"
    case _:
        meaning = "Unknown"
"""
color = "yellow"
# Your code here (use if/elif/else)
if color == 'yellow':
    meaning = 'Wait'
elif color == 'red':
    meaning = 'Stop'
elif color == 'green':
    meaning = 'Go'
else:
    meaning = 'Unknown'
# ------------------------
assert meaning == "Wait"
print("Question 24 passed!")
```
